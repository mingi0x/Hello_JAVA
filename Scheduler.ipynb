```python
import random

date_list={}
yesterday_duty=[]

for date, info in exclusions.items():
    # 1. 날짜 타입 확인 (평일이면 1점, 휴일이면 2점)
    weight = 2.0 if info["type"] == "휴일" else 1.0

    # 2. 해당 날짜 열외자 명단 가져오기
    today_excluded = info["excluded"]

    # 3. 근무 가능자 필터링
    available = [m for m in all_members if m['name'] not in today_excluded and m not in yesterday_duty]

    # 4. 점수가 낮은 순으로 정렬 후 추출
    available.sort(key=lambda x: x['score'])
    # 1. 점수 낮은 상위 12명을 먼저 확정해서 변수에 담기
    current_duty = available[:12]

    # 2. 그 변수(리스트)의 순서를 무작위로 뒤섞기
    random.shuffle(current_duty)
    yesterday_duty=current_duty

    date_list[date] = []

    # 5. 선발된 인원에게 해당 날짜의 가중치(점수) 부여
    for p in yesterday_duty:
        p['score'] += weight
        snapshot = {
            "name": p["name"],
            "rank": p["rank"],
            "score_at_time": p["score"] # 근무 투입 직전 점수 기록
        }
        date_list[date].append(snapshot)
```

```python
import pandas as pd
from openpyxl.utils import get_column_letter

def generate_excel_schedule(date_list, filename="부대_타임라인_근무표.xlsx"):
    with pd.ExcelWriter(filename, engine='openpyxl') as writer:

        for date, members in date_list.items():
            rows = []
            for i in range(0, 12, 2):
                group_num = (i // 2) + 1
                m1 = members[i]
                m2 = members[i+1]

                h = 8 + (i // 2 * 2)
                nh = (20 + (i // 2 * 2)) % 24

                # 조별 4개 슬롯 생성
                for slot in range(4):
                    # 일단 모든 행에 조 이름을 넣습니다 (나중에 병합하기 위함)
                    display_group = f"{group_num}조"

                    # 주간/야간 시간 계산
                    current_h = h if slot < 2 else h + 1
                    current_nh = nh if slot < 2 else (nh + 1) % 24

                    if slot % 2 == 0: # 1, 3번째 타임 (m1)
                        rows.append([
                            display_group,
                            f"{current_h:02d}:00-{current_h:02d}:30", f"{m1['rank']} {m1['name']}",
                            f"{current_nh:02d}:00-{current_nh:02d}:30", f"{m1['rank']} {m1['name']}"
                        ])
                    else: # 2, 4번째 타임 (m2)
                        rows.append([
                            display_group,
                            f"{current_h:02d}:30-{current_h+1:02d}:00", f"{m2['rank']} {m2['name']}",
                            f"{current_nh:02d}:30-{(current_nh+1)%24:02d}:00", f"{m2['rank']} {m2['name']}"
                        ])

                # 조 사이 구분용 빈 줄 (병합 대상에서 제외됨)
                rows.append(["", "", "", "", ""])

            # 데이터프레임 생성 및 시트에 기록
            df = pd.DataFrame(rows, columns=["조 구분", "주간 시간대", "주간 근무자", "야간 시간대", "야간 근무자"])
            df.to_excel(writer, sheet_name=date, index=False)

            # --- 여기서부터 셀 병합 로직 ---
            workbook = writer.book
            worksheet = writer.sheets[date]

            # 1조부터 6조까지 (각 조는 5행씩 차지: 데이터 4줄 + 빈 줄 1줄)
            for g in range(6):
                start_row = 2 + (g * 5) # 엑셀은 1부터 시작하고 헤더가 1행이므로 2부터
                end_row = start_row + 3 # 4개의 행을 병합

                # A열(조 구분) 병합
                worksheet.merge_cells(f"A{start_row}:A{end_row}")

                # 병합된 셀의 정렬을 중앙으로 설정
                cell = worksheet[f"A{start_row}"]
                from openpyxl.styles import Alignment
                cell.alignment = Alignment(horizontal='center', vertical='center')

    print(f"✅ {filename} 생성 완료! '조 구분' 칸이 예쁘게 병합되었습니다.")

# 실행
generate_excel_schedule(date_list)
```

```python
import pandas as pd
import datetime
from pathlib import Path

def get_current_status(base_date_str, target_date_str):
    # 날짜 형식 설정
    date_format = "%Y-%m-%d"

    # 기준일(3월 23일)과 확인할 날짜 설정
    base_date = datetime.datetime.strptime(base_date_str, date_format)
    target_date = datetime.datetime.strptime(target_date_str, date_format)

    # 두 날짜의 차이 계산 (일 단위)
    delta_days = (target_date - base_date).days

    # 몇 주가 지났는지 계산 (7일로 나눔)
    weeks_passed = delta_days // 7

    # 2주 단위로 순환 (0, 1, 0, 1...)
    # 0이면 'a', 1이면 'b'
    cycle = (weeks_passed // 2) % 2

    return "여단 CCTV" if cycle == 0 else "대대 CCTV"

DOW_MAP = {0:"월",1:"화",2:"수",3:"목",4:"금",5:"토",6:"일"}

CSS = """
* { margin:0; padding:0; box-sizing:border-box; }
body {
    font-family: 'Malgun Gothic', sans-serif;
    background: #fff;
    color: #000;
    font-size: 13px;
}
.page {
    width: 210mm;
    padding: 16mm 14mm 14mm;
    page-break-after: always;
}
.page:last-child { page-break-after: auto; }
.doc-title {
    text-align: center;
    border-bottom: 3px solid #000;
    padding-bottom: 10px;
    margin-bottom: 16px;
}
.doc-title h1 { font-size: 17px; font-weight: 700; letter-spacing: 1.5px; }
.doc-title .sub { font-size: 11px; color: #555; margin-top: 4px; }
table { width: 100%; border-collapse: collapse; font-size: 12px; }
thead tr { background: #1a1a1a; color: #fff; }
thead th { padding: 7px 5px; font-weight: 600; text-align: center; font-size: 11.5px; }
.th-day   { background: #3a3a3a; }
.th-night { background: #2a2a5a; }
td { padding: 5px 7px; border: 1px solid #ddd; }
.col-jo {
    width: 50px; text-align: center; font-weight: 700;
    font-size: 13px; background: #f5f5f5; vertical-align: middle;
}
.col-time { width: 95px; text-align: center; font-size: 11.5px; color: #333; font-weight: 500; }
.col-name { font-size: 12.5px; font-weight: 600; padding-left: 12px; }
.day-col   { background: #fffdf5; }
.night-col { background: #f5f5ff; }
tr.jo-border td { border-top: 2px solid #777; }
"""

def parse_excel(filepath):
    xls = pd.read_excel(filepath, sheet_name=None, header=None)
    result = {}
    for sheet_name, df in xls.items():
        df = df.iloc[1:].reset_index(drop=True)
        jos, current_jo, day_slots, night_slots = [], None, [], []
        for _, row in df.iterrows():
            jo_val, day_t, day_n, night_t, night_n = row[0], row[1], row[2], row[3], row[4]
            if pd.isna(day_t):
                if current_jo is not None and day_slots:
                    jos.append({"jo": current_jo, "day": day_slots, "night": night_slots})
                current_jo, day_slots, night_slots = None, [], []
                continue
            if not pd.isna(jo_val):
                current_jo = int(str(jo_val).replace('조', '').strip())
            day_slots.append({"t": str(day_t).strip(), "n": str(day_n).strip()})
            night_slots.append({"t": str(night_t).strip(), "n": str(night_n).strip()})
        if current_jo is not None and day_slots:
            jos.append({"jo": current_jo, "day": day_slots, "night": night_slots})
        result[sheet_name] = jos
    return result

def fmt_title(date_str):
    d = datetime.date.fromisoformat(date_str)
    return f"{d.year}년 {d.month:02d}월 {d.day:02d}일({DOW_MAP[d.weekday()]}) 경계 작전 명령서({get_current_status("2026-03-23",date_str)})"

def build_page(date_str, jos):
    rows = ""
    for ji, j in enumerate(jos):
        span = len(j["day"])
        for i, s in enumerate(j["day"]):
            is_first = i == 0
            border_cls = "jo-border" if (ji > 0 and is_first) else ""
            jo_cell = f'<td class="col-jo" rowspan="{span}">{j["jo"]}조</td>' if is_first else ""
            rows += f"""
            <tr class="{border_cls}">
              {jo_cell}
              <td class="col-time day-col">{s['t']}</td>
              <td class="col-name day-col">{s['n']}</td>
              <td class="col-time night-col">{j['night'][i]['t']}</td>
              <td class="col-name night-col">{j['night'][i]['n']}</td>
            </tr>"""
    return f"""
  <div class="page">
    <div class="doc-title">
      <h1>{fmt_title(date_str)}</h1>
      <div class="sub">주간 08:00 ~ 20:00 &nbsp;|&nbsp; 야간 20:00 ~ 익일 08:00</div>
    </div>
    <table>
      <thead>
        <tr>
          <th style="width:50px">조</th>
          <th class="th-day" style="width:95px">주간 시간대</th>
          <th class="th-day">주간 근무자</th>
          <th class="th-night" style="width:95px">야간 시간대</th>
          <th class="th-night">야간 근무자</th>
        </tr>
      </thead>
      <tbody>{rows}</tbody>
    </table>
  </div>"""

def build_html(pages_html):
    return f"""<!DOCTYPE html>
<html lang="ko">
<head>
<meta charset="UTF-8">
<style>{CSS}</style>
</head>
<body>{pages_html}</body>
</html>"""

import asyncio
import nest_asyncio
nest_asyncio.apply()

def save_pdf(html_str, out_path):
    from playwright.async_api import async_playwright

    async def _run():
        async with async_playwright() as p:
            browser = await p.chromium.launch()
            page = await browser.new_page()
            await page.set_content(html_str, wait_until="networkidle")
            await page.pdf(
                path=out_path,
                format="A4",
                print_background=True,
                margin={"top":"0","right":"0","bottom":"0","left":"0"}
            )
            await browser.close()

    asyncio.get_event_loop().run_until_complete(_run())


print("함수 정의 완료 ✓")
```

```python
EXCEL_PATH='/content/부대_타임라인_근무표.xlsx'

from google.colab import files as colab_files

schedule = parse_excel(EXCEL_PATH)
stem = Path(EXCEL_PATH).stem
print(f"총 {len(schedule)}일치 데이터\n")

generated = []
for date_str, jos in schedule.items():
    out_path = f"{stem}_{date_str}.pdf"
    save_pdf(build_html(build_page(date_str, jos)), out_path)
    generated.append(out_path)
    print(f"  ✓ {out_path}")

all_pages = "".join(build_page(d, j) for d, j in schedule.items())
full_path = f"{stem}_전체.pdf"
save_pdf(build_html(all_pages), full_path)
print(f"\n  ✓ {full_path} (합본)\n")

for f in generated:
    colab_files.download(f)
colab_files.download(full_path)
print("다운로드 완료!")
```
